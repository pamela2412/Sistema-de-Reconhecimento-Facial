```python
import cv2
from ultralytics import YOLO
import matplotlib.pyplot as plt

# Cargar el modelo YOLO entrenado

!pip install roboflow

from roboflow import Roboflow
rf = Roboflow(api_key="NlyCrH9fjWzp0ILAkhLQ")
project = rf.workspace("proyecto-de-deep-learning-dcdm5").project("deteccion-personajes")
version = project.version(1)
dataset = version.download("yolov8")

from ultralytics import YOLO

# Cargar modelo YOLOv8
model = YOLO("yolov8n.pt")  # Usar yolov8n (nano), yolov8s (small), etc.

# Entrenamiento
results = model.train(
    data="/content/Detección-Personajes-1/data.yaml",
    epochs=10,
    imgsz=640,
    batch=16,
    device="0"  # Usar GPU si está disponible
)


```

```python
# Leer imagen
imagePath = '/content/imagen_2025-02-17_155312711.png'
img = cv2.imread(imagePath)
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Detección de caras con YOLO
results = model(img)

for result in results:
    boxes = result.boxes.xyxy  # Coordenadas de las cajas [x1, y1, x2, y2]
    confs = result.boxes.conf   # Confianzas
    cls = result.boxes.cls     # Clases

    for box, conf, class_id in zip(boxes, confs, cls):
        x1, y1, x2, y2 = box.tolist()
        label = model.names[int(class_id)]
        print(f"Objeto: {label}, Confianza: {conf:.2f}, Coordenadas: ({x1}, {y1}, {x2}, {y2})")
```

```python
# Leer imagen
imagePath = '/content/imagen_2025-02-17_181414361.png'
image = cv2.imread(imagePath)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convertir a RGB para matplotlib

# 3. Realizar inferencia
results = model(image)



### Método 2: Dibujar manualmente con OpenCV
for result in results:
    boxes = result.boxes.xyxy.cpu().numpy()
    classes = result.boxes.cls.cpu().numpy()
    confs = result.boxes.conf.cpu().numpy()

    for box, cls, conf in zip(boxes, classes, confs):
        x1, y1, x2, y2 = map(int, box)

        # Dibujar caja
        cv2.rectangle(image_rgb, (x1, y1), (x2, y2), (0, 255, 0), 2)

        # Crear etiqueta
        label = f"{model.names[int(cls)]} {conf:.2f}"

        # Dibujar fondo para el texto
        (w, h), _ = cv2.getTextSize(label, cv2.FONT_HERSHEY_SIMPLEX, 0.6, 1)
        cv2.rectangle(image_rgb, (x1, y1 - 20), (x1 + w, y1), (0, 255, 0), -1)

        # Dibujar texto
        cv2.putText(image_rgb, label, (x1, y1 - 5),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 0, 0), 1)

plt.figure(figsize=(12, 8))
plt.imshow(image_rgb)
plt.axis("off")
plt.show()
```
